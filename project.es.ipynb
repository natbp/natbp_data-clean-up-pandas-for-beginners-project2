#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
max_price = ds["price"].idxmax()

direccion = ds.loc[max_price, "address"]
precio = ds.loc[max_price, "price"]
print(f"La casa con dirección en {direccion} es la más cara y su precio es de {precio} USD.")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
ds_filtrado = ds[ds["price"] > 0]
min_price = ds_filtrado["price"].idxmin()

direccion = ds_filtrado.loc[min_price, "address"]
precio = ds_filtrado.loc[min_price, "price"]
print(f"La casa con dirección en {direccion} es la más barata y su precio es de {precio} USD.")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios
precios_arroyomolinos = ds[ds["level5"] == "Arroyomolinos (Madrid)"]["price"]

plt.hist(precios_arroyomolinos, bins=15)
plt.title("Histograma de precios en Arroyomolinos (Madrid)")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

En el grafico se evidencia que la mayoría de los apartamentos oscilan un valor de 300.000 USD, convirtiéndose este el rango de precios más frecuente.
De igual manera se observa que existen otros dos grupos donde hay una concentración significativa de viviendas en un rango de precios siendo estos entre 200.000 USD y 280.000 USD, y entre 300.000 USD y 370.000 USD.
También se ve que hay muy pocas viviendas con precios inferiores a 100.000 USD, lo que quiere decir que los precios bajos no son comunes en la población de Arroyomolinos.
Así mismo, aparecen algunos valores aislados en los rangos más altos, entre 420.000 USD aproximadamente y 600.000 USD, que representan viviendas muchísimo más costosas donde la mayoría de ese rango oscilan entre los 460.000 USD y los 540.000 USD más o menos.
